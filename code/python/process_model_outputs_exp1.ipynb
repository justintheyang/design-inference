```python
intention_agent = ['1agent', '2agent', 'either_agents']
intention_recipe = ['onion', 'tomato', 'either_recipe']

model_agent = ['1agent', '2agent']
model_recipe = ['onion', 'tomato']

n_seed = 20
```
